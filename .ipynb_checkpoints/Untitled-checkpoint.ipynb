# Getting the data

```python
train_df=pd.read_csv("data/train.csv")
test_df=pd.read_csv("data/test.csv")
val_df=pd.read_csv("data/val.csv")
```

# split data

```python
def split_data(data):
    x=data.drop(columns="price")
    y=data.price.copy()
    return (x,y)
```

```python
X_train,y_train=split_data(train_df)
```

```python
X_test,y_test=split_data(test_df)
```

```python
X_val,y_val=split_data(val_df)
```

# info


```python
X_train.info()
```

```python
#we have to change the dtype of data of journey and arrival_time and desitnation
dt_cols=["date_of_journey","dep_time","arrival_time"]

num_cols=["duration","total_stops"]

cat_cols=[col for col in X_train.columns if(col not in dt_cols) and (col not in num_cols)]
```

```python
cat_cols
```

```python
#we impute them with 

num_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="median")),
    ("scaler",StandardScaler())
])

cat_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("scaler",OneHotEncoder(sparse_output=False,drop="first"))
])

date_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("extracter",DatetimeFeatures(features_to_extract=["month","week","day_of_week","day_of_month"],format="mixed")),
    ("scaler",StandardScaler())
])
time_pipe=Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("extracter",DatetimeFeatures(features_to_extract=["hour","minute"],format="mixed")),
    ("scaler",StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols),
    ("date", date_pipe, ["date_of_journey"]),
    ("time", time_pipe, ["dep_time", "arrival_time"])
])
```

```python
preprocessor.fit_transform(X_train)# we are seeing all rows because of this pd.set_option("display.max_columns",None)
```

# model selection


```python
algorithums={
    "Linear Regression":LinearRegression(),
    "SVM":SVR(),
    "Random Forest":RandomForestRegressor(n_estimators=10),
    "XG Boost":XGBRegressor(n_estimators=10)
}
```

```python
data=pd.concat([train_df,val_df],axis=0)

X_data,y_data=split_data(data)
```

```python
def plot_curves(sizes, mean_scores, std_scores, label, ax):
	ax.plot(
		sizes,
		mean_scores,
		marker="o",
		label=label
	)

	ax.fill_between(
		x=sizes,
		y1=mean_scores - std_scores,
		y2=mean_scores + std_scores,
		alpha=0.5
	)
```

```python
def plot_learning_curves(name, algorithm, figsize=(12, 4)):
	model = Pipeline(steps=[
		("pre", preprocessor),
		("alg", algorithm)
	])
    
    try:
        train_sizes, train_scores, test_scores = learning_curve(
            estimator=model,
            X=X_data,
            y=y_data,
            cv=3,
            scoring="r2",
            n_jobs=1,  # Changed from -1 to 1
            random_state=42
        )
        
        mean_train_scores = np.mean(train_scores, axis=1)
        std_train_scores = np.std(train_scores, axis=1)
        train_score = f"{mean_train_scores[-1]:.2f} +/- {std_train_scores[-1]:.2f}"
        
        mean_test_scores = np.mean(test_scores, axis=1)
        std_test_scores = np.std(test_scores, axis=1)
        test_score = f"{mean_test_scores[-1]:.2f} +/- {std_test_scores[-1]:.2f}"
        
        fig, ax = plt.subplots(figsize=figsize)
        
        plot_curves(
            train_sizes,
            mean_train_scores,
            std_train_scores,
            label=f"Train ({train_score})",
            ax=ax
        )
        plot_curves(
            train_sizes,
            mean_test_scores,
            std_test_scores,
            label=f"Test ({test_score})",
            ax=ax
        )
        
        ax.set(xlabel="Training set sizes", ylabel="R-square", title=name)
        ax.legend()
        
        plt.show()
    
    except Exception as e:
        print(f"An error occurred for {name}: {e}")
```

```python
for name, alg in algorithums.items():
	plot_learning_curves(name, alg)
```
